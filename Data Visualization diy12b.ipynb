## Data Visualization using Matplotlib

<b>Q:</b> Load the "Netflix_titles.csv" data into a DataFrame and plot the following insights

using the Matplotlib library:

1. How much of the Netflix content are TV shows or Movies?
2. In which counrty is Netflix the most popular?

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
netflix_df = pd.read_csv("netflix_titles.csv")
```

```python
netflix_df.head()
```

```python
# 1. how much of the netflix content are TV shows or Movies?
```

```python
# taking types of netflix content
netflix_types = list(set(netflix_df.type))

netflix_type_count=[]
for type in netflix_types:
    netflix_type_count.append(len(netflix_df[netflix_df.type==type]))
plt.title("Type of Netflix Content")
min_content=netflix_type_count.index(min(netflix_type_count))
netflix_explode=np.zeros(len(netflix_type_count))
netflix_explode[min_content]=0.1
plt.pie(netflix_type_count,labels=netflix_types,autopct='%0.2f',explode=netflix_explode)
plt.show()
```

```python
# 2. In which country is Netflix the most popular
```

```python
netflix_df = netflix_df.dropna(subset=["country"])
netflix_countries = list(set(netflix_df.country))

netflix_country_count = []
for country in netflix_countries:
    netflix_country_count.append(len(netflix_df[netflix_df.country==country]))
plt.title("Netflix viewers across the Globe")
plt.pie(netflix_country_count,labels=netflix_countries)
plt.show()
```

<b>Conclusion: as per pie chart, United States is the country in which Netflix is most popular</b>

## Data Visualization using Seaborn

<b>Q:</b> Load the "starbucks_drinkMenu_expanded.csv" data into a DataFrame and find out which "Beverage_category" has highest calories;

plot the insights using a bar plot with the help of the Seaborn library.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
starbucks_df = pd.read_csv("starbucks_drinkMenu_expanded.csv")
```

```python
starbucks_df.head()
```

```python
plt.title("Which of the starbucks drinks have highest calories?")
sns.barplot(x='Calories',y='Beverage_category',data=starbucks_df)
```

<b>Conclusion:</b> As from bar plot <b>Smoothies</b> is the Starbucks drink with highest calories
