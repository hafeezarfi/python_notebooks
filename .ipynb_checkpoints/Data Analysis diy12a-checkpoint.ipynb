## Sales Data Analysis - I

Q: Load the "sample_superstore.csv" dataset provided on LMS into a DataFrame, using the Matplotlib library,

plot the following insights from  the dataset:

1. Compare the total sales for each category of products
2. Number of deals in each state of America

<b>Note:</b> You can use any plot of your choice to show the insights

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

sample = pd.read_csv("SampleSuperStore.csv")
```

```python
sample.head()
```

```python
print("1. Compare the total sales for each category of products")
sample_total_sales=dict()
for i in set(sample['Category']):
    sample_total_sales[i]=sample[sample['Category']==i].Sales.sum()
plt.title("Total Sales for each Category of Products")
plt.xlabel("Categories")
plt.ylabel("Sales")
plt.bar(sample_total_sales.keys(),sample_total_sales.values())
plt.show()
```

```python
print("2. Number of deals in each state of America")

sample_deals_states=dict()
for i in set(sample['State']):
    sample_deals_states[i]=sample[sample['State']==i].Sales.count()
plt.title("Total Number of Deals in Each American State")
plt.xlabel("States")
plt.ylabel("Deals")
plt.xticks(rotation='vertical')
plt.bar(sample_deals_states.keys(),sample_deals_states.values())
plt.show()
```
