```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


cereal_df = pd.read_csv('cereal.csv')
```

```python
cereal_df.head()
```

```python
# 1. info() and describe() functions
cereal_info = cereal_df.info()
```

```python
cereal_description = cereal_df.describe()
```

```python
cereal_description
```

```python
# 2. Replace manufacturer names with the dictionary of given names
manufacturer_dict = {
    'A': 'American Home Food Products',
    'G': 'General Mills',
    'K': 'Kelloggs',
    'N': 'Nabisco',
    'P': 'Post',
    'Q': 'Quaker Oats',
    'R': 'Ralston Purina'
}

careal_df['mfr']=cereal_df['mfr'].map(manufacturer_dict)

cereal_df.head()
```

```python
# 3 Find out top 5 cereal manufacturers based on customer ratings
top_manufacturers = cereal_df.groupby('mfr')['rating'].mean().sort_values(ascending=False)
```

```python
top_manufacturers.head()
```

```python
# 4. barplot
sns.set(style="whitegrid")
plt.figure(figsize=(10,6))
sns.barplot(x=top_manufacturers.index,y=top_manufacturers.values, palette="viridis")
plt.title("Manufacturers with highest customer ratings")
plt.xlabel("Manufacturer")
plt.ylabel("Customer Rating")
plt.xticks(rotation=45)
plt.show()
```

```python
# 5. pie chart of manufacturers that are largest producers of cereals
producer_counts=cereal_df['mfr'].value_counts()

plt.figure(figsize=(8,8))
plt.pie(producer_counts,labels=producer_counts.index, autopct="%1.1f%%",colors=sns.color_palette("viridis",len(producer_counts)))
plt.title("Manufacturers which are largest Producers")
plt.show()
```
